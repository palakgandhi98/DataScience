```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#define the parameter of the pareto distribution

alpha = 3
xm = 1

#create an array of x values

x= np.linspace(0.1,10,1000)

#calculate the y values of Pareto Distribution
y = alpha * (xm**alpha) / (x**(alpha+1))

plt.plot(x,y)
```

#log-log plot for pareto distribution

```python
#create the log-log plot for verify the distribution is pareto distribution

plt.plot(np.log(x),np.log(y))

#add labels and titles

plt.xlabel('X')
plt.ylabel('P(x)')
plt.title('Pareto Distribution')
```

#Q-Q plot for pareto distribution

```python
import numpy as np
import scipy.stats as st
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#define the parameters for Pareto Distribution

alpha = 2
xm = 1

#Generate the random for pareto distribution

x = st.pareto.rvs(b=alpha,scale=xm,size=1000)
```

```python
plt.hist(x)
```

```python
sns.histplot(x)
```

```python
#fit pareto distribution to the data

params = st.pareto.fit(x,floc=0)

#create a pareto distribution object with the fittied parameters

dist = st.pareto(b=params[0],scale=params[2])
```

```python
#Create the QQ plot of the data using the Pareto Distribution

fig = sm.qqplot(x,dist=dist,line='45')

#Add labels and title to the plot
plt.title('QQ Plot of Pareto Distribution With Fit Pareto ')
plt.xlabel('Theoretical Quantiles')
plt.ylabel('Sample Quantiles')
```
