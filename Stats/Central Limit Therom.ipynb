```python
import numpy as np
import matplotlib.pyplot as plt

#set the parameter
num_samples = 1000
sample_size = 50

#Poisson Distribution Parameter
poisson_lambda = 5

#Gamma Distribution Parameter
gamma_shape = 2
gamma_scale= 1

#Binomial Distribution Parameter
binomial_n = 10
binomial_p=0.5

#Generate sample from a Uniform Distribution
poisson_sample = np.random.poisson(lam=poisson_lambda,size=(num_samples,sample_size))
gamma_sample = np.random.gamma(shape=gamma_shape,scale=gamma_scale,size=(num_samples,sample_size))
binomail_sample = np.random.binomial(n=binomial_n,p=binomial_p,size=(num_samples,sample_size))


#calculate the sample mean
poisson_mean = np.mean(poisson_sample, axis=1)
gamma_mean = np.mean(gamma_sample, axis=1)
binomail_mean = np.mean(binomail_sample, axis=1)

#before mean
#plot the histrogram of sample mean
fig, axs = plt.subplots(3,1,figsize=(8,15))

#poisson distribution
axs[0].hist(poisson_sample,bins=30,density=True)
axs[0].set_title('Histrogram Of Poisson Distribution Samples')
axs[0].set_xlabel('Poisson Sample')
axs[0].set_ylabel('Density')

#gamma distribution
axs[1].hist(gamma_sample,bins=30,density=True)
axs[1].set_title('Histrogram Of Samples (Gamma Distribution)')
axs[1].set_xlabel('Gamma Sample')
axs[1].set_ylabel('Density')

#Binomial distribution
axs[2].hist(binomail_sample,bins=30,density=True)
axs[2].set_title('Histrogram Of Sample (Binomial Distribution)')
axs[2].set_xlabel('Poisson Sample')
axs[2].set_ylabel('Density')


plt.show()

#after applying CLT
#plot the histrogram of sample mean
fig, axs = plt.subplots(3,1,figsize=(8,15))

#poisson distribution
axs[0].hist(poisson_mean,bins=30,density=True,edgecolor='black')
axs[0].set_title('Histrogram Of Sample Mean (Poisson Distribution)')
axs[0].set_xlabel('Poisson Mean')
axs[0].set_ylabel('Density')

#gamma distribution
axs[1].hist(gamma_mean,bins=30,density=True,edgecolor='black')
axs[1].set_title('Histrogram Of Sample Mean (Gamma Distribution)')
axs[1].set_xlabel('Gamma Mean')
axs[1].set_ylabel('Density')

#Binomial distribution
axs[2].hist(binomail_mean,bins=30,density=True,edgecolor='black')
axs[2].set_title('Histrogram Of Sample Mean (Binomial Distribution)')
axs[2].set_xlabel('Poisson Mean')
axs[2].set_ylabel('Density')


plt.show()
```

#Prove Cental Limit Theorem
#Pop mean   sample mean x and
#pop variance 2  sample variance 2/n

```python
import numpy as np

#set parameter
num_samples = 10000
sample_size = 50

#Gamma Distribution Paramters
gamma_shape = 2
gamma_scale = 1

#calculate the theoretical mean and variance
theoretical_mean = gamma_shape * gamma_scale
theoretical_variance = (gamma_shape * gamma_scale **2)

#Generate the samples from Gamma Distribution
gamma_samples = np.random.gamma(shape=gamma_shape,scale=gamma_scale, size=(num_samples,sample_size))

#camculate the samples mean
gamma_samples_mean = np.mean(gamma_samples, axis=1)

#Calculate the Empirical mean and variance of the gamma samples
empirical_mean = np.mean(gamma_samples_mean)
empirical_variance = np.var(gamma_samples_mean)

#Comapare the theoretical and empirical Values

print(f'Theoretical Mean: {theoretical_mean:.4f}')
print(f'Empirical Mean: {empirical_mean:.4f}')
print(f'\n')
print(f'Theoretical Variance: {theoretical_variance:.4f}')
print(f'Emprical Mean: {empirical_variance:.4f}')
```

```python
import numpy as np

# Set the parameters
population_size = 100000
sample_size = 50
num_samples = 100

# Generate a random representative sample of salaries (in thousands)
# You should replace this with actual collected salary data
np.random.seed(42)  # Setting a seed for reproducibility
population_salaries = np.random.lognormal(mean=4.5, sigma=0.8, size=population_size)

# Generate multiple samples and calculate the sample means and standard deviations
sample_means = []
sample_std_devs = []

for _ in range(num_samples):
  sample_salaries = np.random.choice(population_salaries, size=sample_size)
  sample_means.append(np.mean(sample_salaries))
  sample_std_devs.append(np.std(sample_salaries))

# Calculate the average of the sample means and the standard error
average_sample_means = np.mean(sample_means)
standard_error = np.std(sample_means) / np.sqrt(num_samples)

# Calculate the 95% confidence interval
margin_of_error = 1.96 * standard_error
lower_limit = average_sample_means - margin_of_error
upper_limit = average_sample_means + margin_of_error

# Report the results
print(f"Estimated average salary (in thousands): {average_sample_means:.2f}")
print(f"95% confidence interval (in thousands): ({lower_limit:.2f}, {upper_limit:.2f})")
```
