```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 10 #total number of trails
p = 0.5 # probability of success
size = 1000 #number of samples to generate

binomial_dist = np.random.binomial(n,p,size)
#binomial_dist = np.random.binomial(n,0.2,size)
#binomial_dist = np.random.binomial(n,0.9,size)


plt.hist(binomial_dist,density=True)
plt.show()
```
