```python
from sklearn.datasets import load_diabetes

import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
```

```python
X,y = load_diabetes(return_X_y=True)
```

```python
print(X.shape)
print(y.shape)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
reg = LinearRegression()
reg.fit(X_train,y_train)
```

```python
print(reg.coef_)
print(reg.intercept_)
```

```python
y_pred = reg.predict(X_test)
r2_score(y_test,y_pred)
```

```python
class SGDRegressor:

    def __init__(self,learning_rate=0.01,epochs=100):

        self.coef_ = None
        self.intercept_ = None
        self.lr = learning_rate
        self.epochs = epochs

    def fit(self,X_train,y_train):
        # init your coefs
        self.intercept_ = 0
        self.coef_ = np.ones(X_train.shape[1])

        for i in range(self.epochs):
          for j in range(X_train.shape[0]):
            #Select index randomly from X_train
            idx = np.random.randint(0,X_train.shape[0])

            #update coef_ and intercept_
            #return single number or scalar
            #calculate derivative of select index
            y_hat = np.dot(X_train[idx],self.coef_) + self.intercept_

            #Calculate and update intercept of selected index
            intercept_der = -2 * (y_train[idx] - y_hat)
            self.intercept_ = self.intercept_ - (self.lr * intercept_der)

            #Calculate and update coeifecent of selected index
            coef_der = -2 * np.dot((y_train[idx] - y_hat),X_train[idx])
            self.coef_ = self.coef_ - (self.lr * coef_der)

        print(self.intercept_,self.coef_)

    def predict(self,X_test):
        return np.dot(X_test,self.coef_) + self.intercept_
```

```python
sgd = SGDRegressor(epochs=50,learning_rate=0.5)
```

```python
import time
start = time.time()
sgd.fit(X_train,y_train)
print("The time taken is",time.time() - start)
```

```python
y_pred = sgd.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```
