```python
#fib

def fib(n):
  if n==1 or n==0:
    return 1
  else:
    return fib(n-1) + fib(n-2)
```

```python
fib(2)
```

```python
fib(4)
```

```python
fib(25)
```

```python
fib(40)
```

```python
#fib memoization

def fib(n,d):
  if n in d:
    return d[n]
  else:
    d[n] = fib(n-1,d) + fib(n-2,d)
    return d[n]
```

```python
d = {0:1,1:1}
fib(40,d)
```

```python
fib(50,d)
```
