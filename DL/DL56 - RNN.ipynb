```python
from keras import Sequential
from keras.layers import Dense, SimpleRNN
```

```python
model = Sequential()

model.add(SimpleRNN(3, input_shape=(4,5)))
model.add(Dense(1, activation='sigmoid'))

model.summary()
```

```python
print(model.get_weights()[0].shape) #input --> hidden 5*3
model.get_weights()[0]
```

```python
print(model.get_weights()[1].shape) # 3 * 3 recurrent
model.get_weights()[1]
```

```python
print(model.get_weights()[2].shape) #bias node
model.get_weights()[2]
```

```python
print(model.get_weights()[3].shape) #hidden --> Output node
model.get_weights()[3]
```

```python
print(model.get_weights()[4].shape) #output node output y_cap or bias
model.get_weights()[4]
```
