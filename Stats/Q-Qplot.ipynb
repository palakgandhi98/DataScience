#Uniform Plot Check

```python
import numpy as np
import scipy.stats as st
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
#Generate set of random data

x = np.random.uniform(low=0,high=1,size=1000)
```

```python
plt.hist(x)
```

```python
#fit a uniform distribuation to the data

params = st.uniform.fit(x)
dist = st.uniform(loc=params[0],scale=params[1])
```

```python
#create a QQ plot of the data using the uniform distribution
fig = sm.qqplot(x, dist=dist, line='45')

#add titles and labels for Q-Q plot
plt.title('Q-Q Plt of Uniform distribution with uniform fit')
plt.xlabel('Theoretical Quantiles')
plt.ylabel('Sample Quantiles')

#Show the plot
plt.show()
```
